# Magnets: How do they work? Exploring the 2D Ising Model

Magnetism on a microscopic level is caused by electrons in the atom "shells". The Ising model is a intuitive but powerful way of describing such a system. In this notebook, we'll first learn about the Ising model and try to get some intuition for it. To look into the physical consequences, we'll solve it numerically with my c++ solver magneto. Everything on [github.com/s9w/magneto](https://github.com/s9w/magneto). Finally, we'll learn about the algorithms that are used to solve it and their differences and peculiarities.

Quantum Mechanics tells us that electrons have a property called spin which is in one of two states, usually called "up" and "down". In **Ising model**, that is represented by $\sigma=+1$ and $\sigma=-1$, with $\sigma$ just being the standard symbol for spin states. The spins are arranged in a two or three dimensional square lattice. We'll look at two dimensions here.

![Alt text](doc/ising.svg)

Each electron spin creates a magnetic field which interacts with other electrons. In good approximation, it's enough to only look at the interaction between next neighbor spins. So each spin only is affected by its four direct adjacent sites.

The behaviour of all physical systems boils down to minimizing the overall energy. So let's try to reason about the energy of such a system: An electron spin produces either a positive or negative magnetic field. But it's also influenced by such fields, both external and those of the other spins. A positive field will push an electron spin state towards a positive alignment and vice versa. So it's intuitive that without external fields and disturbances, a configuration where many spins are parallel (in the same state) is energetically beneficial. Put into an equation, the energy of the whole system (without an external magnetic field) is given by the Hamiltonian:

$$H = -J\sum_{\langle ij\rangle } \sigma_i \sigma_j$$

That way of writing a sum just means a summation over next neighbours for every single spin. We'll assume that the coupling constant $J$ is 1 for the moment, more on that later. For good results, a large grid is crucial. To help with "finite size effects", periodic boundary conditions are used. So for a 2x2 grid with either all positive or all negative spins, the energy would be -8. Usually the energy is divided by the number of sites in the grid so that the it's in a nice range between -2 and 0.

So we know what energy a specific configuration has. But a square grid with length $L$ has $L^2$ places, which amounts to $2^{(L^2)}$ possible states. Even with conservative grid sizes that's a large number. Which of those configurations are actually physically realized and how often? More precise: What's their probability distribution for different temperatures? Statistical mechanics has the answer: The Boltzmann distribution

$$\exp\left(-\frac{H}{T}\right)$$

Normalized, it looks like this:

```python
ax = plt.subplot(xlabel="Energy", ylabel="Probability", title="Boltzmann distribution")
E = np.linspace(0, 4.5, num=50)

for i, T in enumerate([0.8, 1.5, 3.5]):
    ax.plot(E, 1.0/T*np.exp(-E/T), "-", label="T={}".format(T), color=colors3[i])
ax.legend();
```

As expected, configurations with the lowest total energy are always the most probable. But at high temperatures, the distribution becomes flatter and states with higher energies become more likely. Therefore we expect the average total energy to increase with temperature.

Physical interpretation: A higher temperature increases thermal energy and therefore thermal fluctuations. They compete with the magnetic interaction between the spins. While the magnetic forces will push the system towards a uniform state, thermal effects have a random nature. With high T, the thermal effects get stronger and eventually dominate over the spin interactions.

Now we got everything we need. To recap: Ising model says we can write such a system as grids with $\pm 1$; the Hamiltonian gives the relationship between a spin configuration and its energy; the Boltzmann distribution gives us a probability for each energy. So when we have algorithms to solve the ising model, we get a stream of configurations that obey physics and on which we can observe physical behaviour. The algorithms to generate the configurations efficiently are desribed at the bottom of this notebook.

To visualize how the system looks at different temperatures, we'll run a first simulation. The two spin states are represented with white and black. Note that all simulation inputs are included, but commented out because some of them run quite long.

```python
# !magneto -N1=50 -N2=0 -TMin=1.5 -TMax=3.5 -L=109 -TSteps=4 -states=states
fig = plt.figure(figsize=(8,2))
for i in range(4):
    with open("states{}.txt".format(i)) as f:
        T = float(f.readline())
    ising_grids = np.loadtxt("states{}.txt".format(i), delimiter=",", skiprows=1)
    ax = fig.add_subplot(1, 4, i+1)
    ax.grid(False)
    ax.imshow(ising_grids, cmap=plt.cm.Greys, interpolation="none")
    ax.set_title("T={:.2f}".format(T))
fig.tight_layout()
```

As expected! At low temperatures, there is little thermal energy and the magnetic spin interaction dominates. This results in mostly uniform states. With rising temperatures, the thermal fluctuations become stronger and eventually turn the system into random noise.

## Energy, heat capacity and phase transitions
To get into physical details, we'll first look at the **energy**. It was defined with the hamiltonian above.

```python
# !magneto -L=30 -TSteps=15 -en=en -dist=normal -alg2=metro
Tc = 2.0/np.log(1+np.sqrt(2))
ax = plt.subplot(111, xlabel="T/T_c", ylabel="Energy", ylim=(-2,0), xlim=(0,2))
T, E = np.loadtxt("en.txt", delimiter=", ", unpack=True)
ax.plot(T/Tc, E);
```

The energy increases with temperature, as expected. Here, the temperature on the x-axis was divided by a special temperature $T_c$ called the **critical** or **Curie temperature**. That's because a lot of things happen with systems like this at that point.

To see that, we'll look at the (specific) **heat capacity** of the system. It's the amount of heat added to a system, divided by the temperature change. That sounds odd, but is quote straight forward: Metal has a high capacity, so it'll barely heat up if you add a certain amount of heat. But something like air will!

There are two ways of calculating $C_V$. One is simply the temperature derivation of energy. The other is from the energy variation.

$$C_V=\frac{1}{T^2}(\langle E^2\rangle - \langle E\rangle^2)=\frac{dE}{dT}$$

Let's look at both:

```python
# !magneto -L=30 -TSteps=12 -cv=cv -dist=normal -alg2=metro
ax = plt.subplot(xlabel="T/T_c", ylabel="Heat capacity")

T, cv = np.loadtxt("cv.txt", delimiter=", ", unpack=True)
ax.plot(T/Tc, cv, label="by variation")
T, E = np.loadtxt("en.txt", delimiter=", ", unpack=True)
ax.plot(T[:-1]/Tc, np.diff(E)/np.diff(T), label="by derivation")
ax.legend(loc="upper right");
```

We see that both ways yield equal results up to a factor and numerics. The heat capacity has a peak right at $T_c$. It's different for all kinds of systems and even in the Ising model it differs between 2D and 3D. For the 2D case it's known exactly: $$T_c=\frac{2}{\ln(1+\sqrt 2)}\approx 2.269$$

## Magnetization, Susceptibility and Correlation
The energy increases with temperature, as expected. Here, the temperature on the x-axis was divided by a special temperature $T_c$ called the **critical** or **Curie temperature**. That's because a lot of things happen with systems like this at that point. It becomes clearer when we look at the **magnetization** of the total system, which is just the sum over all spins. It causes the magnetic field of the total system.

Systems with all negative or all positive spins are physically identical since their energy is the same. So usually, the absolute value of the spins is used to calculate magnetization. That makes the plots prettier and the calculations easier. So the magnetization is $\lvert m \rvert=\lvert \sum \sigma_i \rvert$.

If we recall how the systems looked, it's easy to predict that magnetization should be at its maximum for low, and vanish at high temperatures. Let's see:

```python
# !magneto -L=30 -TSteps=12 -N2=200 -N3=10 -mag=mag -dist=normal -alg2=metro
ax = plt.subplot(xlabel="T/T_C", ylabel="Magnetization |m|", xlim=(0,2), ylim=(0,1))
T, m = np.loadtxt("mag.txt",  delimiter=", ", unpack=True)
ax.plot(T/Tc, m);
```

Bingo! And now it's easy to see why $T_c$ is so important. It's the temperature where the magnetization of the system becomes zero. More generally, the system undergoes a **phase transition**, a huge and important concept in physics. Usually there are two types of phase transitions: "First order" ones, for example the boiling and evaporating of water. And "second order" or continuous transitions like this system, which is actually a model for the transition between the **ferromagnetic** and **paramagnetic** phase.

**Ferromagnetism** means that when an external magnetic field is applied, the electron spins lign up with it and cause themselves a big magnetization. This only works at temperatures below the critical point. We saw that at low temperatures, the spins are mostly uniform and that mostly positive or negative states are energetically identical and therefore equally probable. When even a small external field is applied, they'll easily align to it.

Above $T_c$, magnetization is heavily disturbed by thermal fluctuations. An external field can still "force" the spins to align, but this **paramagnetism** is typically much weaker than ferromagnetism. Common magnets you can buy are therefore all in the ferromagnetic phase. They would lose their strength above their Curie temperature, which is quite high for quantum reasons.

## Correlation and susceptibility
The **correlation** $G(i,k)$ between two spins $\sigma_i$ and $\sigma_k$ is given by the correlation function

$$G(i,k) = \langle\sigma_i \sigma_k\rangle - \langle\sigma_i\rangle \langle\sigma_k\rangle $$

It decreases with their distance $d$ and can be modeled by an exponential function:

$$G(i)=\sum_k G(i,k)=\exp(-\frac d \xi)$$

That **correlation length** $\xi$ can be interpreted as average size of clusters or as typical range of spin fluctuations. It's also a way to calculate the **magnetic susceptibility** $\chi$ with $\chi \sim \frac 1 T \xi$. The susceptibility can also be calculated from the variation of the magnatization:

$$\chi=\frac{1}{T}(\langle m^2\rangle - \langle m\rangle^2)$$

```python
# !magneto -L=40 -TSteps=18 -N2=500 -chi=ising_chi -dist=normal -alg2=metro
# !magneto -L=40 -TSteps=15 -N2=700 -N3=15 -corr=corr -dist=normal -alg2=metro

ax = plt.subplot(xlabel="T/T_c", ylabel="Magnetic susceptibility $\chi$")
ax2 = ax.twinx()
ax.grid(False)
ax2.grid(False)

T, chi = np.loadtxt("ising_chi.txt",  delimiter=", ", unpack=True)
ax.plot(T/Tc, chi, label="by variation")
T, corrLen = np.loadtxt("corr.txt",  delimiter=", ", unpack=True)
ax2.plot(T/Tc, corrLen, "red", label="by correlation")
    
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines + lines2, labels + labels2, loc="upper right");
```

The susceptibility can be used to look at the critical behaviour with more detail. Not only does it diverge at $T_c$, but like many properties it follows a power law. If we define a "reduced temperature" $\tau=\frac{T-T_c}{T_c}$, then $\chi$ is on both sides of the divergence proportional to $\lvert\tau\rvert^{-\gamma}$. That $\gamma$ is called a **critical exponent**. For our 2D ising case, the theoretical value is 1.75. Let's try to calculate it with a fit.

```python
# !magneto -L=100 -TSteps=15 -N2=500 -N3=10 -chi=chi -TMin=2.7 -TMax=3.5 -alg2=sw -dist=normal
ax = plt.subplot(111, xlabel="Reduced temperature $\\tau$", ylabel="Magnetic susceptibility $\chi$")

# Fit
def chi_fit(tau, gamma, a):
    return a*np.power(tau,-gamma)
T, chi = np.loadtxt("chi.txt", delimiter=", ", unpack=True)
tau = (T-Tc)/Tc
popt, pcov = curve_fit(chi_fit, tau, chi)

# Plot
ax.plot(tau, chi, "o", label="data")
ax.plot(tau, chi_fit(tau, *popt), "--", label="fit")
ax.legend(loc="upper right")

# Result
gamma = popt[0]; gamma0 = 1.75
print("gamma: {:.2f}, deviation: {:.1f}%".format(gamma, 100.0*(gamma-gamma0)/gamma0))
```

This is a difficult computation and it's fun to play around with the different algorithms, grid size, temperature ranges, N2, N3. This run took about 30 seconds on my machine and was within $\pm 10\%$. Other critical exponents can be found similarly.

## Finite size effects
The position of the peaks of the heat capacity or the susceptibility drift away from the theoretical value for grids with finite size. That drift can be used to extrapolate the ideal value for $L\to\infty$. For that, the critical temperature is plotted against the inverse grid size. The intersection of the fit with the $\frac 1 L =0$ line yields a much better value of $T_C$ than any of the direct computations.

```python
# !magneto -L=8  -TSteps=9 -TMin=2.2 -TMax=2.5 -avgN=9500  -measure=cv -o=ising_cv_8  -threads=4
# !magneto -L=10 -TSteps=9 -TMin=2.2 -TMax=2.5 -avgN=15000 -measure=cv -o=ising_cv_10 -threads=4
# !magneto -L=12 -TSteps=9 -TMin=2.2 -TMax=2.5 -avgN=12000 -measure=cv -o=ising_cv_12 -threads=4
# !magneto -L=16 -TSteps=9 -TMin=2.2 -TMax=2.4 -avgN=9500  -measure=cv -o=ising_cv_16 -threads=4
# !magneto -L=32 -TSteps=9 -TMin=2.2 -TMax=2.4 -avgN=6500  -measure=cv -o=ising_cv_32 -threads=4
```

```python
# Plot critical temps vs inverse grid sizes
grid_sizes = np.array([8,10,12,16,32])
T_critical_array = []
for L in grid_sizes:
    cv = np.loadtxt("ising_cv_{}.txt".format(L),  delimiter=", ")
    T_critical = cv[np.argmax(cv[:,1],axis=0)][0]
    T_critical_array.append(T_critical)
L_inv = 1.0/grid_sizes # inverse grid size

fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111, xlabel="1/L", ylabel="Critical temperature", xlim=(0,0.15), ylim=(2.25,2.4))
ax.set_title("Critical temperature for different L")
ax.plot(L_inv, T_critical_array, "o")
ax.set_xticks(np.arange(0,0.16, 0.05))

# Fit with a linear function
def T_fit(x,a,b):
    return a+b*x
popt, pcov = curve_fit(T_fit, L_inv, T_critical_array)
x = np.linspace(0,0.14, 10)
ax.plot(x, T_fit(x, *popt), "--")

# Result
gamma = popt[0]; gamma0 = 1.75


T_c_fit = T_fit(0, *popt)
print("Calculated Tc: {:.3f}, deviation: {:.2f}%".format(T_c_fit, 100.0*(T_c_fit-Tc)/Tc))

ax.axhline(y=Tc, marker="");
```

This fit yielded an extrapolated value of $T_C\approx 2.271$, which deviates from the theoretical value only by $0.06\%$.

## Algorithms
We've seen at the beginning, that for a lattice length of $L$, there are $2^{L\cdot L}$ possible configurations. To do statistics, one could generate all of them and weigh them with the Boltzmann factor $e^{-H/T}$. A much better approach is to generate the system configurations so that they're Boltzmann distributed.

The most popular way to do this is the Metropolis algorithm. It's based on markov chains and uses an equilibrium condition called [detailed balance](http://en.wikipedia.org/wiki/Detailed_balance#Reversible_Markov_chains) to prove a simple assumption: If the system is in a state with energy $E_0$ and we flip the value of a single spin so that the system has now $E_1$, the probability to accept that change should be 

$$p=\min\left(1, \exp\left(-\frac{\Delta E}{T}\right)\right) \quad \text{ with } \Delta E=E_1-E_0$$

to generate the states in a Boltzmann distributed way.

```python
ax = plt.subplot(xlabel="$\Delta E$", ylabel="Probability", ylim=(0,1.1), title="Metropolis acceptance probability")
dE = np.linspace(-2, 2)

for i, T in enumerate([0.8, 1.5, 3.5]):
    ax.plot(dE, np.minimum(1, np.exp(-dE/T)), "-", label="T={}".format(T), color=colors3[i])
ax.legend();
```

That's quite simple and elegant! But it suffers from an effect called critically slowing down, meaning that it becomes increasingly ineffective near the critical temperature. To understand, recall that the system has large clusters of parallel spins there. Those clusters are energetically optimal. And because of their size, a single spin flip like in Metropolis isn't doing much. Let's look at that in detail:

Interestingly, the Metropolis and Swendsen-Wang (SW) Algorithms have different behaviours. Metropolis suffers from an effect called critically slowing down, meaning that it becomes increasingly ineffective near the critical temperature. That problem doesn't exist in cluster algorithms like SW. But while SW is still fast around the critical point, the position is much further off the theoretical value for small grid sizes.

```python
fig = plt.figure(figsize=(6,7))
l1 = [8,16,32]
Tc1 = []

ax = fig.add_subplot(211, xlabel="T/T_c", ylabel="Susceptibility", xlim=(0.8,1.8), title="Metropolis")
for i,L in enumerate(l1):
    T, cv = np.loadtxt("chi_m{}.txt".format(L),  delimiter=", ", unpack=True)
    ax.plot(T/Tc, cv, marker="", label="L={}".format(L), color=colors3[i])
    Tc1.append( T[np.argmax(cv)] )
ax.set_yscale('log')
ax.set_ylim(0.1,100)
ax.legend(loc="upper right")

l2 = [16,32,50]
Tc2 = []
ax = fig.add_subplot(212, xlabel="T/T_c", ylabel="Susceptibility", xlim=(0.8,1.8), title="Swendsen-Wang")
for i,L in enumerate(l2):
    T, cv = np.loadtxt("chi_sw{}.txt".format(L),  delimiter=", ", unpack=True)
    ax.plot(T/Tc, cv, marker="", label="L={}".format(L), color=colors3[i])
    Tc2.append( T[np.argmax(cv)] )
ax.set_yscale('log')
ax.set_ylim(0.1,100)
ax.legend(loc="upper right");
fig.tight_layout()
```

It can be seen that for both algorithms that with increasing system size, the peak height increases and the position converges towards the ideal. The SW peak positions are much further away from the ideal. A plot of peak position against grid size makes this even clearer:

```python
ax = plt.subplot(111, xlabel="L", ylabel="T/Tc")
ax.plot(l1, Tc1/Tc, label="Metr")
ax.plot(l2, Tc2/Tc, label="SW")
ax.legend(loc="lower right");
```
