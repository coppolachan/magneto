# 2D Ising Model

Plotting imports:

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

We're looking at the 2D ising model with an Energy of:
$$H = -\sum_{\langle ij\rangle } \sigma_i \sigma_j$$

The sum is over the next neighbors and the spins $\sigma_i$ have values of $\pm 1$. The system is at Temperature $T$

Starting from a random initial configuration, the system can be evolved with either **Metropolis** or **Swendsen-Wang** Algoritms. They reach an equilibrium for each Temperature. Plotting typical states for increasing temperatures:

```python
# ising -TMin=1.5 -TMax=4.0 -L=80 -TSteps=4 -measure=states -o=ising_states
fig = plt.figure(figsize=(8,2))
for i in range(4):
    ising_grids = np.loadtxt("ising_states{}.txt".format(i),  delimiter=",")
    ax = fig.add_subplot(1, 4, i+1)
    ax.imshow(ising_grids, cmap=plt.cm.Greys, interpolation ="none")
fig.tight_layout()
```

## Energy and Magnetization

Typical properties are the **energy** and the absolute value of the **magnetization** $\lvert m \rvert=\lvert \sum \sigma_i \rvert$. The critical temperature is $$T_c=\frac{2}{\ln(1+\sqrt 2)}\approx 2.269$$

```python
Tc = 2.0/np.log(1+np.sqrt(2))
```

```python
# ising -L=60 -TSteps=9 -avgN=500 -measure=energy -o=ising_e
# ising -L=60 -TSteps=12 -avgN=500 -measure=mag -o=ising_m

fig = plt.figure(figsize=(6,3))

ax = fig.add_subplot(121, xlabel="T/T_c", ylabel="Energy", ylim=(-2,0), xlim=(0,2))
ax.grid(True)
T, E = np.loadtxt("ising_e.txt", delimiter=", ", unpack=True)
ax.plot(T/Tc, E, lw=2)

ax = fig.add_subplot(122, xlabel="T/T_C", ylabel="Magnetization |m|", xlim=(0,2))
ax.grid(True)
T, m = np.loadtxt("ising_m.txt",  delimiter=", ", unpack=True)
ax.plot(T/Tc, m, lw=2)

fig.tight_layout()
```

## Heat capacity

The **heat capacity** $c_v$ can be calculated either from the variation of the energy or from its temperature derivation:
$$c_V=\frac{1}{T^2}(\langle E^2\rangle - \langle E\rangle^2)=\frac{dE}{dT}$$

```python
# ising -L=60 -TSteps=12 -avgN=500 -measure=energy -o=ising_e
# ising -L=60 -TSteps=9 -avgN=500 -measure=cv -o=ising_cv

fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111, xlabel="T/T_c", ylabel="Heat capacity")
ax2 = ax.twinx()
ax.grid(True)
T, cv = np.loadtxt("ising_cv.txt",  delimiter=", ", unpack=True)
ax.plot(T/Tc, cv, label="by variation", lw=2)
T, E = np.loadtxt("ising_e.txt",  delimiter=", ", unpack=True)
ax2.plot(T[:-1]/Tc, np.diff(E), "red", label="by derivation", lw=2)
    
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc="center left", labelspacing=0.0)
fig.tight_layout()
```

## Correlation and susceptibility

The **correlation** $G(i,k)$ between two spins $\sigma_i$ and $\sigma_k$ is given by the correlation function
$$G(i,k) = \langle\sigma_i \sigma_k\rangle - \langle\sigma_i\rangle \langle\sigma_k\rangle $$
It decreases with their distance $d$ and can be modeled by an exponential function:
$$G(i)=\sum_k G(i,k)=\exp(-\frac d \xi)$$
That **correlation length** $\xi$ can be interpreted as average size of clusters or as typical range of spin fluctuations. It's also a way to calculate the **magnetic susceptibility** $\chi$ with $\chi=\frac 1 T \xi$. The susceptibility can also be calculated from the variation of the magnatization:
$$\chi=\frac{1}{T}(\langle m^2\rangle - \langle m\rangle^2)$$

```python
# ising -L=60 -TSteps=9 -avgN=500 -measure=chi -o=ising_chi
# ising -L=60 -TSteps=9 -avgN=500 -measure=corrLen -o=ising_corrLen

fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111, xlabel="T/T_c", ylabel="Magnetic susceptibility $\chi$")
ax2 = ax.twinx()
ax.grid(True)

T, chi = np.loadtxt("ising_chi.txt",  delimiter=", ", unpack=True)
ax.plot(T/Tc, chi, label="by variation", lw=2)
T, corrLen = np.loadtxt("ising_corrlen.txt",  delimiter=", ", unpack=True)
ax2.plot(T/Tc, corrLen, "red", label="by correlation", lw=2)
    
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc="center left", labelspacing=0.0)
fig.tight_layout()
```

The heat capacity as well as the susceptibility both show critical behaviour, as expected.

## Finite Size

The position of the peaks of the heat capacity or the susceptibility drift away from the theoretical value grids of finite size. That drift can be used to extrapolate the ideal value for $L\to\infty$. For that, the critical temperature is plotted against the inverse grid size. The intersection of the fit with the $\frac 1 L =0$ line yields a much better value of $T_C$ than any of the direct computations.

```python
# ising -L=8  -TSteps=9 -TMin=2.2 -TMax=2.5 -avgN=9500  -measure=cv -o=ising_cv_8  -threads=4
# ising -L=10 -TSteps=9 -TMin=2.2 -TMax=2.5 -avgN=15000 -measure=cv -o=ising_cv_10 -threads=4
# ising -L=12 -TSteps=9 -TMin=2.2 -TMax=2.5 -avgN=12000 -measure=cv -o=ising_cv_12 -threads=4
# ising -L=16 -TSteps=9 -TMin=2.2 -TMax=2.4 -avgN=9500  -measure=cv -o=ising_cv_16 -threads=4
# ising -L=32 -TSteps=9 -TMin=2.2 -TMax=2.4 -avgN=6500  -measure=cv -o=ising_cv_32 -threads=4

# Plot critical temps vs inverse grid sizes
grid_sizes = np.array([8,10,12,16,32])
T_critical_array = []
for L in grid_sizes:
    cv = np.loadtxt("ising_cv_{}.txt".format(L),  delimiter=", ")
    T_critical = cv[np.argmax(cv[:,1],axis=0)][0]
    T_critical_array.append(T_critical)
L_inv = 1.0/grid_sizes # inverse grid size

fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111, xlabel="1/L", ylabel="Critical temperature", xlim=(0,0.15), ylim=(2.25,2.4))
ax.grid(True)
ax.plot(L_inv, T_critical_array, "o", lw=2)
ax.set_xticks(np.arange(0,0.16, 0.05))

# Fit with a linear function
from scipy.optimize import curve_fit
def T_fit(x,a,b):
    return a+b*x
popt, pcov = curve_fit(T_fit, L_inv, T_critical_array)
x = np.linspace(0,0.14, 10)
ax.plot(x, T_fit(x, *popt))

T_c_fit = T_fit(0, *popt)
print("Calculated Tc:", T_c_fit)
print("Deviation:", 100.0*(T_c_fit-Tc)/Tc)

# Theoretical value
ax.axhline(y=Tc, lw=2)
```

This fit yielded an extrapolated value of $T_C\approx 2.270$, which deviates from the theoretical value only by $0.0608\%$.
